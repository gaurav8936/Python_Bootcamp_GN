<font size = 8> Interactions between Functions

```python
example = [1,2,3,4,5,6,7]
```

```python
from random import shuffle
# this uses the shuffle function
```

```python
shuffle(example)
# Note that the shuffle function does not return anything
```

```python
result = shuffle(example)
```

```python
result
# This proves that the shuffle function does not return anything even if 
# assigned to a variable
# all the values are in place
```

```python
example
# to see the returned values, you will need to call the origin list
```

***

<font color = red> creating our own function for shuffle that will return a value

```python
def shufflelist(mylist):
    shuffle(mylist)
    return(mylist) # this return is what stores the value of the newly shuffled list
```

```python
result = shufflelist(example)
```

```python
result
# as you can see below, this now returns a list of values of the newely shuffled list
```

<font color = blue> second function to accept user input

```python
def player_guess():
    # note that unlike earlier function, this does not take a variable value
    
    guess=''
    # this sets the starting value to blank
    
    while guess not in ('0', '1', '2'):
        guess = input("Pick a number: 0,1,2")
        # input always returns a string
    
    return int(guess)
```

```python
player_guess()
```

<font color = blue> third function to combine above two functions

```python
def check_guess(mylist1, guess):
    if mylist1[guess] == 'O':
        print("Correct")
    else:
        print("Wrong guess")
        print(mylist1)
    
```

<font color = blue> Putting it all together

```python
# Steps
#1 Initial List
#2 Shuffle List
#3 User guess the location
#4 Check guess

#1 Initial List
mylist = ['','O','']


#2 Shuffle List
mixedup_list = shufflelist(mylist)


#3 User guess the location
guess = player_guess()


#4 Check guess
check_guess(mixedup_list, guess)
```
