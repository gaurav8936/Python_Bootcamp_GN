<font size = 8> Grabbing an Image

```python
import requests

import lxml

import bs4
```

```python
# Wikipedia Commons
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Deep_Blue_(chess_computer)')
```

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
```

```python
soup
```

***

```python
soup.select('img')

# returns everything with an img tag in it
```

```python
soup.select('img')[0]
```

When you go to the url within SRC, you get this




```python
soup.select('.thumbimage')

# this is a class based selection
```

```python
# this is the raw content(binary file) of the image
```

```python
# Python can read this file

# The process to save this is
## create a fil
## save to file
## close the file
```

```python
# create a new jpg file

f = open('my_computer_image.jpg', 'wb') # wb denotes write binary
```

```python
f.write(image_link.content)
```

```python
f.close()
```
