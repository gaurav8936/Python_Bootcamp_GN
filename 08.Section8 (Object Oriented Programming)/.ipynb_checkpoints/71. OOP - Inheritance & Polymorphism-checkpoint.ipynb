```python
# Define the base class

class Animal():
    
    def __init__(self):
        print("animal created")
    
    # METHOD
    def who_am_i(self):
        print("I am an animal")
        
    #METHOD
    def eat(self):
        print("I am eating")
```

```python
my_animal = Animal()
```

```python
my_animal.eat()
```

```python
my_animal.who_am_i()
```

***

```python
# recreate the dog class
```

```python
class Dog(Animal):
    
    # inherit the animal class
    
    # This is now a derived class
    
    def __init__(self):
        Animal.__init__(self)
        
        # This creates an instance of the Animal class while create an instance of the 
        # Animal class
    
        print("Dog created")
```

```python
my_dog = Dog()
```

```python
# all oldmethods are not available
```

```python
my_dog.eat()
```

```python
my_dog.who_am_i()
```

***

```python
# Over writing one of the older methods
```

```python
class Penguin(Animal):
    
    # inherit the animal class
    
    # This is now a derived class
    
    def __init__(self):
        
        Animal.__init__(self)
        
        # This creates an instance of the Animal class while create an instance of the 
        # Animal class
    
        print("Penguin created")
        
    # Now to overwrite an inherited method
    
    def who_am_i(self):
        print("I am a penguin")
```

```python
my_penguin = Penguin()
```

```python
my_penguin.who_am_i()
```

***

```python
# Adding new methods to the inherited class
```

```python
class Bat(Animal):
    
    # inherit the animal class
    
    # This is now a derived class
    
    def __init__(self):
        
        Animal.__init__(self)
        
        # This creates an instance of the Animal class while create an instance of the 
        # Animal class
    
        print("Bat created")
        
    
    # New Method
    
    def makesound(self):
        print("JUUUUUUYTY")
    
    # Now to overwrite an inherited method
    
    def eat(self):
        print("I am a bat and eating")
```

```python
my_bat = Bat()
```

```python
my_bat.eat()

# THis is overwriting an inherited method
```

```python
my_bat.makesound()

# This is not inherited but added on
```

```python
my_bat.who_am_i()

# This is inherited and unchanged
```

# Polymorphism

<font color = red> the use of SAME METHOD name across classes

```python
# refers to the way different object classes can share the same method name

## In the example below, each class (Dog, Cat) has the speak method but returns 
## value unique to the object
```

```python
class Dog():
      
    def __init__(self, name):
        self.name = name
    
    # METHOD
    def speak(self):
        return self.name + " says WOOF"
```

***

```python
### Create a CAT class that is very similar to the Dog class (speak different only)

class Cat():
      
    def __init__(self, name):
        self.name = name
    
    # METHOD
    def speak(self):
        return self.name + " says MEEOOOOOW"
```

```python
niko = Dog("niko")
felix = Cat('felix')
```

```python
print(niko.speak())
```

```python
print(felix.speak())
```

***

```python
## POLYMORPHISM DEMO
```

```python
for pet in [niko, felix]:
    
    print(type(pet))
    print(pet.speak())
```

```python
## Another way of showing this is through a function
# This is the more common way of doing this
```

```python
def pet_speak(pet):
    
    print(pet.speak())
```

```python
pet_speak(niko)
```

```python
pet_speak(felix)
```


***

```python
# The more common method is to use ABSTRACT CLASSED and INHERITENCE
```

```python
# Abstract classes never expect to be INSTANTIATED
# THIS ONLY SERVES AS A BASE CLASS
```

```python
# The ANIMAL class is never expected to be instantiated
# This is an abstract class
# This will only be a base class

class Animal():
    
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        raise NotImplementedError("Subclass must implement this abstract method")
    
    # This section will ensure you cannot create an instance of Animal
```

```python
myanimal = Animal("Fred")
```

```python
myanimal.speak()
```

<font color = red> It expects that other classes will inherit the Animal class and over write the speak method

***

```python
class Dog(Animal):
    
    def speak(self):
        return self.name + "  says WOOF!"
```

```python
class Cat(Animal):
    
    def speak(self):
        return self.name + "  says MEOOOOF!"
```

```python
fido = Dog("Fido")
```

```python
gemma = Cat("Gemma")
```

```python
print(fido.speak())
```

```python
print(gemma.speak())
```
