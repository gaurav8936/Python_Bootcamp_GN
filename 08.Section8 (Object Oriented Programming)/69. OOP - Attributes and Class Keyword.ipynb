```python
class Dog():
    
    # a function inside a class is called a METHOD
    
    def __init__(self, breed):
        
        self.breed = breed
        
        # breed is a parameter
```

```python
# create a variable

my_dog = Dog( breed = 'Lab')
```

```python
type(my_dog)

# shows the Dog class
```

```python
my_dog.breed
```

```python
# This instance of the Dog class is Lab
```
