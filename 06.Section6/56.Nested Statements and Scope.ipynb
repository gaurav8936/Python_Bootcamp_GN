```python
# example of enclosing function local

name = 'THIS IS A GLOBAL STRING'

def greet():
    # this is the enclosing function of def hello()
    name = 'Sammy'
    # overriding the name variable
    
    def hello():
        print('Hello '+ name)
    hello()
```

```python
greet()
```

```python
# example of global function

name = 'THIS IS A GLOBAL STRING'

def greet():
    # this is the enclosing function of def hello()
#    name = 'Sammy'
    # overriding the name variable
    
    def hello():
        print('Hello '+ name)
    hello()
```

```python
greet()
```

```python
# GlOBAL

name = 'THIS IS A GLOBAL STRING'

def greet():
    # ENCLOSING
    name = 'Sammy'

    
    def hello():
        # LOCAL
        name = 'I am a local'
        print('Hello '+ name)
    hello()
```

```python
greet()
```

```python
## REASSIGN X LOCALLY

x = 50

def func(x):
    print(f'X is {x}')

    # LOCAL REASSIGNMENT
    x = 200
    print(f'I just locally changed X to {x}')
```

```python
func(x)
```

```python
print(x)

# This returns 50 becuase the reassignment is happening only insdie the local namespace
# It is not going to change anything at a higher scope
```

<font color = red> What if you really wanted to change the x value globally

```python
# Change x value outside the local 

x = 50

#def func(x):
def func():
    global x # This declares a global X
    print(f'the current local value of X is {x}')
    
    # LOCAL REASSIGNMENT ON A GLOBAL VAIRIABLE
    x = 'NEW VALUE'
    print(f'I just changed the GLOBAL value of x to {x}')
```

```python
## Chack the initial value of X

print(x)
```

```python
## Run the function

func()
```

```python
## Print the new value of X

print(x)
```

```python
x = 50

def func(x):
    # REMEBER TO TAKE THE X is as a PARAMETER (this wasnt teh case with global)
    
    print(f'X is {x}')
    
    # LOCAL REASSIGNMENT ON A GLOBAL VARIABLE
    
    x = 'NEW VALUE'
    print(f'I just locally changed global X to {x}')
    
    return x
    # This step is also difference (as we are returning the value of x)
    
    
```

```python
print(x)

# Iitial value
```
