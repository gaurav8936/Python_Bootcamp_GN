## String Interpolation

```python
print('This is a string{}'.format('INSERTED'))
```

## what the .fomat does is look at the {} and then insert the content

```python
print("How are you doing {}".format('Mahroushka'))
```

```python
print("What is the time{}".format('Bobbin'))
```

```python
print('The {} {} {}'.format('fox', 'brown', 'lazy'))
```

```python
print('The {2} {0} {1}'.format('fox', "brown", "lazy"))
```

```python
print('The {} {} {}'.format ('fox', 'brown', 'lazy'))
```

```python
print('The {f} {b} {l}'.format(f='fox', b='brown', l='lazy'))
```

```python
print('I would like to take the time to introduce {s} to {p} today!'.format(s='John', 
      p = 'Philadelphia'))
```

```python
result = 100/777
```

```python
result
```

```python
print("The result was {}".format(result))
```

```python
print ("The result was {r}".format(r = result))
```

<font size = 20> formatting "{value:width,precision f}" <font>

```python
print("The result was {r:1.3f}".format(r=result))
```

```python
print("The result was {c:1.3f}".format(c = .9943333))
```

```python
print("The result was {c:10.10f}".format(c = .9943333))
```
