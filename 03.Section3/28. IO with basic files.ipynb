## Read Mode

```python
%%writefile myfile3.txt
Hello this is a text file
This is the second line
This is the third line
This is the fourth line
```

```python
with open('myfile3.txt', mode = 'r') as f:
    print(f.read())
```

## Append

```python
with open('myfile3.txt', mode = 'a') as f:
    f.write('This is the fifth line')
```

```python
with open('myfile3.txt', mode = 'r') as f:
    print(f.read())
```

```python
with open('myfile3.txt', mode = 'a') as f:
    f.write('\nThis is the sixth line')
```

```python
with open('myfile3.txt', mode = 'r') as f:
    print(f.read())
```

## Write (with either overwrite or create a new file

```python
with open('myfile4_write', mode = 'w') as f:
    f.write('I created this file')
```

```python
with open('myfile4_write', mode = 'r') as f:
    print(f.read())
```
