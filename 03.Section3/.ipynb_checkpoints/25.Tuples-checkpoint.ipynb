### Tuples can have different types (like strings, integers ..)

```python
t = ('one', 2)
```

## Tuples can also be sliced and diced

```python
t[0]
```

```python
t[-1]
```

## There are two basic methods for Tuples -- INDEX, COUNT

```python
t = ('a', 'a', 'b')
```

```python
t.count('a')
```

```python
t.index('a')
```

` index returns the very first time the value appeared - in this case at location 0

```python
t.index('b')
```
