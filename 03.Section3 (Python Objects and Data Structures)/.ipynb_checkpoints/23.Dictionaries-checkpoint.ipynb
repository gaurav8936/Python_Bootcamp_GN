Dictionaries are key value pairs that allows storing of unordered data

### Dictionaries cannot be sorted -- all you need is the key value pair

Python determines where to most efficiently insert the key value pairs

```python
my_dictionary = {'key1' : 'value1', 'key2' : 'value2'}
```

```python
my_dictionary
```

```python
my_dictionary['key1']
```

```python
prices_lookup = {'apple' : 2.99, 'apricot' : 1.22, 'oranges' : '0.98', 'milk' : 5.80}
```

```python
prices_lookup['apple']
```

### --- see that we did not have to specify the location in order to get the result

## Dictionaries can hold other Dictionaries or Lists

```python
d = {'k1' : 123, 'k2' : [0,1,2], 'k3' : {'insidekey' : 100}}
```

` this is an example of a dictionary with an embedded list as well as another dictionary `

```python
d['k2']
```

```python
d['k3']
```

below is an example of extracting from an embedded dictionary

```python
d['k3']['insidekey']
```

below is an example of extracting from an embedded list

```python
d['k2'] [2]
```

### example of replacing one item with an uppercase

```python
d = {'key1': ['a', 'b', 'c']}
```

```python
d
```

```python
mylist = d['key1']
```

```python
mylist
```

```python
letter = mylist[2]
```

```python
letter
```

```python
letter.upper()
```

` however, this took very many steps `

```python
d['key1']
```

```python
d['key1'][2]
```

```python
d['key1'][2].upper()
```
