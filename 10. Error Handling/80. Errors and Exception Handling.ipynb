## Refer Exceptions Documentation from python.org

```python
try:
    f = open('testfile', 'w')
    f.write("Write a test line to this file")
except TypeError:
    # This is misleading
    # Except -- implies only if there is a TypeError
    print("There is a TypeError")
except OSError:
    print('Hey you have an OS Error')
finally:
    # Will always execute no matter run
    print("I always run")
```

***

```python
# Let us induce an error by changing w to r

try:
    f2 = open('testfile', 'r')
    f2.write("Write a test line to this file")
except TypeError:
    # This is misleading
    # Except -- implies only if there is a TypeError
    print("There is a TypeError")
except OSError:
    print('Hey you have an OS Error')
except:
    print('All other exceptions')
    # This is a catch all
finally:
    # Will always execute no matter run
    print("I always run")
```

```python
def ask_for_int():
    try:
        result = int(input("Please provide number:  "))
    except: 
        print("Whoops! This is not a number")
    finally:
        print("End of try/execpt/finally")
```

```python
ask_for_int()
```

### Remember a WHILE TRUE loop always needs a BREAK statement

```python
# within a while loop
# The else/bvreak statement is tired directly to the while loop only
# finally is after the while loop executes or break

def ask_for_help2():
    
    while True:
        try:
            result = int(input("Please provide a number: "))
        except:
            print("Whoops! This is not a number")
            continue
        else:
            print("Yes thank you")
            break
        finally:
#             print("End of try/except/finally")
#             print("I will always run at the end")
            print("I am going to ask you again! \n")
```

```python
ask_for_help2()
```
