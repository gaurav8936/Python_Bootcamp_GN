### Zip file built into Python

```python
# Create two files
```

```python
f = open('fileone.txt', 'w+')
f.write('One File')
f.close()
```

```python
f = open('filetwo.txt', 'w+')
f.write('One File')
f.close()
```

***

```python
import zipfile

#built into python
```

```python
# First created a zip file

comp_file = zipfile.ZipFile('comp_file.zip', 'w')

# notice the ZipFile -- upper adn lower case
```

***

```python
# compress text files and insert into zip file
```

```python
comp_file.write('fileone.txt', compress_type = zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('filetwo.txt', compress_type = zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

***

```python
# extract items from zip file
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip', 'r')
```

```python
zip_obj.extractall('extracted_content')
```

```python
# we have now taken the contents of the comp_file.zip file and created a new file
```
