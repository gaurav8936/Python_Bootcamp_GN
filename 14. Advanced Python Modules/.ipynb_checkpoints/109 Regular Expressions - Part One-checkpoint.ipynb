```python
text = "The agent's phone number is 408-555-1234. Call soon!"
```

```python
'phone' in text
```

```python
import re
```

```python
pattern = 'phone'
```

```python
re.search(pattern, text)
```

```python
# reports not just the mathc of the word phone, but also index (span)
```

***

```python
pattern2 = 'Not in text'
```

```python
re.search(pattern2, text)

# This does not return anything because the text is not there
```

***

```python
pattern3 = 'phone'
```

```python
re.search(pattern3, text)
```

```python
match = re.search(pattern3, text)

# save to match object
```

```python
match
```

```python
# now you can call the match obect


match.span()

# this gives us the position
```

```python
# start

match.start()
```

```python
# end

match.end()
```

***

```python
# if you have multiple matches, you just get back the first one
```

```python
text2 = "My phone once, my phone twice, my phone thrice"
```

```python
match = re.search('phone', text2)
```

```python
match
```

```python
# that only gave us the first instance
```

***

```python
# find all functions
```

```python
matches = re.findall('phone', text2)
```

```python
matches
```

```python
len(matches)
```

***

```python
# Combining both SEARCH (which returns specs) and FINDALL (which returns all strings)
```

```python
for match in re.finditer('phone', text2):
    
    print(match)
```

```python
# Just getting spans

for match in re.finditer('phone', text2):
    
    print(match.span())
```

```python
# Returning actual text you are matching for

for match in re.finditer('phone', text2):
    
    print(match.group())
```
