## os.walk  Directory Tree Generator -- most useful

```python
# Directory tree

# relevant for tuple unpacking
```

```python
os.getcwd()
```

```python
file_path = '/Users/ganarasi/Dropbox/Berkeley2022/11.Python/\
31.Python_Bootcamp_GN/14. Advanced Python Modules/Example_Top_Level'
```

```python
# Start with tuple unpacking
# start working with the example top level folder

for folder, sub_folders, files in os.walk(file_path):
    
    print(f"currently looking at {folder}")
    print('\n')
    print('The subfolders are: ')
    for sub_fold in sub_folders:
        print(f"\t Subfolder: {sub_fold}")
    
    print('\n')
    print("The files are: ")
    for f in files:
        print(f"\t File: {f}")
    print('\n')
```
