* counterpart of \w
* \W\W\W\W\W

## Quantifiers

+

* occurs one ore more times
* Version \w-\w+
* Version A-b1_1 ------------ here the 1 occurs more than once

{3}

* occurs exactly 3 times
* D{3}  ---------- D is for non digit fyi
* abc
* ie a non digit appears 3 times

{2,3}

* Occurs 2-4 times
* \d{2,4}
* 123


{3,}

* occurs 3 or more times
* \w{3,}
* anycharacters

{*}

* occurs zero or more times
* A *B* C*
* AAAACCC

#### Find phone numbers and extract area codes

<font size = 4> Compile Function

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results = re.search(phone_pattern, text)
```

```python
results.group()
```

```python
results.group(1)

# Remember that group ordering starts at 1 (unlike 0 for the rest of python)
```

```python
results.group(2)
```

```python
results.group(4)

# no such group
```
