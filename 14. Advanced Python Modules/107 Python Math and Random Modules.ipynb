# RANDOM Module

***

```python
# COMPUTERS USE PSEUDO RANDOM GENERATORS

# TEHRE IS METHOD TO THE MADNESS OF GENERATING RANDOM NUMBERS
```

```python
# RANDOM SEED 

# allows us to get the same set of random numbers over and over again
```

***

```python
import random
```

```python
# Random integers

random.randint(0,100)
```

***

```python
# Testing with the same set of random numbers

# SEED
```

```python
# SEED AND RANDINT NEED OT BE IN THE SAME CELL

random.seed(101)

random.randint(0,100)

# common use is 42 which is the hitchhikers guide to the galaxy
```

```python
# you always get back 74
```

***

```python
# Do this again

random.seed(101)

random.randint(0,100)
```

```python
# Now just run randint

random.randint(0,100)
```

```python
# Now just run randint

random.randint(0,100)
```

```python
# the three values 74, 24 and 69 keep repeating
```

***

```python
# redo hte experiment differently

random.seed(101)

print(random.randint(0,100)) # expect 74
print(random.randint(0,100)) # expect 24
print(random.randint(0,100)) # expect 69
print(random.randint(0,100))
print(random.randint(0,100))
```

```python
# setting the SEED value is what is the root cause
```

***

```python
# taking a random item from the list
```

```python
mylist = list(range(0,20))
```

```python
mylist
```

```python
# grab a random integer from this list
```

```python
random.choice(mylist)
```

```python
random.choice(mylist)
```

***

```python
# Grabbing multiple items
```

```python
# Sample with or without replacement
```

```python
# Sample with replacement

# This allows the same integer to be selected more than once
```

```python
random.choices(population = mylist, k=10)

# this will randomly pick numbers
```

```python
# You see that 7 is selected twice
```

***

```python
# Sampling WITHOUT replacement
```

```python
# once you choose a number, you cant pick it again
```

```python
random.sample(population=mylist, k=10)
```

```python
# items will never be repeated here
```
