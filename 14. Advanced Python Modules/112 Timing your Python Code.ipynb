```python
# function 1
```

```python
# Using List Comprehension

def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(10)
```

***

```python
# Function 2

# Using Mapping Function
```

```python
def func_two(n):
    return list(map(str, range(n)))
```

```python
func_two(10)
```

***

```python
import time
```

```python
# Current Time before

start_time = time.time()

# Run Code

# result = func_one(10000000) ## just have a large number

result = func_two(10000000)

# Current Time after

end_time = time.time()

# Difference

elapsed_time = end_time - start_time

print(elapsed_time)
```

# TIMEIT Module

```python
# this has higher precision
```

```python
import timeit
```

```python
stmt = '''
func_one(100)
'''
```

```python
# What code needs to be run BEFORE STMT

setup = '''
def func_one(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt, setup, number = 1000000)
```

***

```python
# Do the same thing for func_two
```

```python
stmt2 = '''
func_two(100)
'''
```

```python
setup2 = '''
def func_two(n):
    return list(map(str, range(n)))
'''
```

```python
timeit.timeit(stmt2, setup2, number = 1000000)
```
