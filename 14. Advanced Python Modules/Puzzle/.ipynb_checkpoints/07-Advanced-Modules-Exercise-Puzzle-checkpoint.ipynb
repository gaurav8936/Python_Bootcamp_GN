```python
import re
```

```python
# create pattern

pattern = r'\d{3}-\d{3}-\d{4}'
```

```python
# test the pattern

test_string = 'here is a phone number 123-123-1234'
```

```python
re.findall(pattern, test_string)
```

```python
# the pattern works
```

***

```python
# Let us make this into a function that opens a text file and searches for the pattern
```

```python
# using search instead of findall

# CREATE A SEARCH FUNCTION

def mysearch(file, pattern = r'\d{3}-\d{3}-\d{4}'):
    f = open(file, 'r')
    text = f.read()
    
    if re.search(pattern, text):
        return re.search(pattern, text)
    else:
        return ''
```

***

```python
# traverse all files using the OS module
```

```python
import os
```

```python
results = []
```

```python
# get the current working directory & the folder 'extracted_content' within it

os.getcwd() +'/extracted_content/extracted_content' 
```

```python
for folder, subfolders, files in os.walk(os.getcwd() +'/extracted_content/extracted_content'):
    
    for f in files:
        full_path = folder+'/'+f
        
        results.append(mysearch(full_path))
```

```python
results
```

```python
for r in results:
    if r != '':
        print(r.group())
```
