` print only even numbers `

```python
for num3 in mylist1:
    #check for even
    if num3%2 == 0:
        print(num3)    
```

```python
for num4 in mylist1: ## had to restart kernel for the variable mylist1
    #check for even
    if num4%2 == 0:
        print(num4)
    else:
        print(f'Odd Number: {num4}')
```

` keep a running tally `

```python
list_sum1 = 0 ## This sets the counter to 0

for num2 in mylist1:
    list_sum1 = list_sum1 + num2
print(list_sum1)
```

```python
list_sum1 = 0 

for num2 in mylist1:
    list_sum1 = list_sum1 + num2
    ## indentation
    print(list_sum1) ## note the print is inside the for loop and runs for each iteration
```

## Tuple pairs in a LIST

<font color = 'red'> special property of for loops with tuples. Note TUPLE UNPACKING

```python
mylist2 = [(1,2), (3,4), (5,6), (7,8)]
```

```python
len(mylist2)
```

```python
for item in mylist2:
    print(item)
```

<font size = 20, color = 'red'> Tuple Unpacking

```python
for (a,b) in mylist2:
    print(a)
    print(b)
```

```python
for (a,b) in mylist2:
    print(a)
#    print(b)
```

` technically you don't need paranthesis as well `

```python
for a,b in mylist2:
    print (a)
    print(b)
```

```python
mylist3 = [(1,2,3), (5,6,7), (8,9,10)]
```

# Dictionary

```python
d1 = {'k1': 1, 'k2': 2, 'k3':3}

for item1 in d1:
    print(item1)
```

<font color = 'red'> Note you only iterate through the keys, not the values

```python
d2 = {'k1':1, 'k2':2, 'k3':3}

for item2 in d2.items():     ## d2.items() allows you to access both key & value in a dictionary
    print(item2)
```

<font size = 20, color = 'red'> You can use the same unpacking technique as with tuples

```python
d3 = {'k1':1, 'k2':2, 'k3':3}

for key1, value1 in d3.items():
    print(key1)
    print(value1)
```

```python
## to only print values, you could also use .value() instead of .item()

d4 = {'k1':1, 'k2':2, 'k3':3}

for value2 in d4.values():
    print(value2)
```

` Note: Dictionaries are UNORDERED `
