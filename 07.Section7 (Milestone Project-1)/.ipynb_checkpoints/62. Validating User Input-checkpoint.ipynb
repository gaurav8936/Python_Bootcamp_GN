```python
def user_choice():
    
    choice = 'WRONG'
    
    # You start by assuming the condition is false (ie there is no digit)
    
    while choice.isdigit() == False:
    
    # The loop keeps running for any non-digit value of choice
    # So the user will be ocntinuously promoted to enter a digit
    
        choice = input("Please enter a number (0-10):" )
    
    return int(choice)
```

```python
# Adding error message

def user_choice2():
    
    choice = 'WRONG'
    
    # You start by assuming the condition is false (ie there is no digit)
    
    while choice.isdigit() == False:
    
    # The loop keeps running for any non-digit value of choice
    # So the user will be ocntinuously promoted to enter a digit
    
        choice = input("Please enter a number (0-10):" )
        
        if choice.isdigit() == False:
            print("Sorry that is not a digit")
    
    return int(choice)
```

```python
def user_choice3():
    
    # VARIABLES
    
      
    # INITIAL CHOICE
        
    choice = 'WRONG'
    acceptabe_range = range(0,10)
    within_range = False
    
    # TWO CONDITIONS TO CHECK
    # DIGIT CHECK
    # WITHIN RANGE IS FALSE

    
    while choice.isdigit() == False or within_range == False:
    
        choice = input("Please enter a number (0-10):" )
        
        # DIGIT CHECK
        if choice.isdigit() == False:
            print("Sorry that is not a digit")
        
        # RANGE CHECK
        
        if choice.isdigit() == True:   # can also be wrfitten as if choice.isdigit():
        
        ## notice also that there is no ELIF
        
            if int(choice) in acceptable_values:
                within_range = True
            else:
                print('Sorry you are out of acceptable range')
                within_range = False
            
    
    return int(choice)
```
