` The use cases below is to create a list by iterating through a FOR loop `

```python
mystring1 = "hello"
```

```python
mylist1 = [] ## empty list
for letter1 in mystring1:
    mylist1.append(letter1)
```

```python
mylist1
```

```python
mylist2 = [letter2 for letter2 in mystring1]
```

```python
mylist2
```

<font color = 'red'> this is a flattened out for loop

```python
mylist3 = [x3 for x3 in 'word']
```

```python
mylist3
```

```python
mylist4 = [x4 for x4 in range(0,11)]  ## runs through the range
mylist4
```

## functions inside the loop

```python
mylist5 = [num5**2 for num5 in range(0,11)]  ## appends the square to the list
mylist5
```

```python
mylist6 = [num6 for num6 in range(1,25) if num6%2 ==0]  ## this only prints out even 
## numbers in the range
mylist6
```

```python
mylist7 = [num7**2 for num7 in range(1,25) if num7%2 == 0] ## this prints the square of the 
## even numbers
mylist7
```

### Complex example of converting celcius to farenheit

<font color = 'red'> List comprehension is hard to read (one liner) -- so it is more readable to follow code like the for loop which are longer, but easier

```python
celcius8 = [0,10,20,34.5]

farenheit8 = [((9/5)*temp8 + 32) for temp8 in celcius8]

farenheit8
```

` Doing the same thing in a for loop `

```python
celcius9 = [0,10,20,34.5]

farenheit9 = []
for temp9 in celcius9:
    farenheit9.append((9/5)*temp9+32)

farenheit9
```

## If Else Statements -- order is different

```python
result10 = [x10 if x10%2 == 0 else 'ODD' for x10 in range(0,11)] ### the order for FOR 
## is different -- this is confusing

result10
```

## Nested Loops

<font color = 'red'> Below is a solution for nested loops using simpler for loops

```python
mylist11 = []

for x11 in [2,4,6]:
    for y11 in [25,33,66]:
        mylist11.append(x11*y11)

mylist11
```

<font color = 'red'> Avoid the list comprehension solution where possible -- this is unreadable later

```python
mylist12 = [x12*y12 for x12 in [2,4,6] for y12 in [1,10,100]]

mylist12
```
