```python
if True:
    print("It is true!")
```

```python
if 3>2:
    print("it is true!")
```

```python
hungry = True
if hungry:
    print("feed me")
```

```python
hungry1 = False
if hungry1:
    print("feed me")
```

```python
hungry2 = False
if hungry2:
    print("feed me")
else:
    print("I am not hungry")
```

## All 3, if, elif, else

```python
loc = "bank"
if loc == "Auto Shop":                      ## see the ==
    print("Auto shops are cool!")
else:
    print("I do not much about autos")
    
    
```

```python
loc = "bank1"
if loc == "Auto Shop":                      ## see the ==
    print("Auto shops are cool!")
elif loc == "bank1":
    print("money is cool!")
elif loc == "Store":
    print("amazon is great")
else:
    print("I do not much about autos")
```

```python
name = "Jose"
if name == "Frankie":
    print("Hello Frankie")
elif name == "Sammy":
    print("Hello Sammie")
else:
    print("What is your name?")
```
